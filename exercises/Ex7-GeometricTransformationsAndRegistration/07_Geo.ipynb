# 07 - Geometric transformations and landmarks

```python
import numpy as np
import cv2
import sklearn
from skimage import color, io, measure, img_as_ubyte
from skimage.measure import profile_line
from skimage.transform import rescale, resize, rotate, EuclideanTransform, warp, AffineTransform, SimilarityTransform, swirl, matrix_transform
from skimage.util import img_as_float, img_as_ubyte
from skimage.filters import threshold_otsu
import matplotlib.pyplot as plt
import math
```

```python
def show_comparison(original, transformed, transformed_name):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(8, 4), sharex=True,
                                   sharey=True)
    ax1.imshow(original)
    ax1.set_title('Original')
    ax1.axis('off')
    ax2.imshow(transformed)
    ax2.set_title(transformed_name)
    ax2.axis('off')
    io.show()
```

## Image rotation

### Exercise 1 - Rotate

```python
in_dir = 'data/'
im_org = io.imread(f'{in_dir}NusaPenida.png')

# angle in degrees - counter clockwise
rotation_angle = 10
rotated_img = rotate(im_org, rotation_angle)
show_comparison(im_org, rotated_img, "Rotated image")
```

```python
rot_center = [0, 0]
rotated_img = rotate(im_org, rotation_angle, center=rot_center)
show_comparison(im_org, rotated_img, "Rotated image with [0,0] center")
```

### Exercise 2 - Center points and background

```python
rotated_img = rotate(im_org, rotation_angle, mode="reflect")
show_comparison(im_org, rotated_img, "Rotated image with reflect mode")
```

```python
warped_img = warp(rotated_img)
show_comparison(im_org, warped_img_img, "Rotated image with reflect mode")
```
